```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("https://raw.githubusercontent.com/aku23004/fifa/main/fifa_data.csv")
data2 = data1[['Name','Age', 'Nationality', 'Club', 'Release Clause']]

print("Example of data objects: Each row represents a data object:")
data1.head()
```

```python
# Display every row
data2.head(pd.set_option('display.max_rows', None))
```

```python
# View random samples
sample = data2.sample(3)

# Display the random sample
print(sample)
```

```python
# Total number of rows and columns.
num_rows, num_columns = data2.shape

print(f'Total number of rows: {num_rows}')
print(f'Total number of columns: {num_columns}')
```

```python
data2['Release Clause'] = pd.to_numeric(data2['Release Clause'].replace('[\€,]', '', regex=True)
                                           .replace('K', 'e3', regex=True)
                                           .replace('M', 'e6', regex=True),
                                       errors='coerce')
```

```python
# Check for any NULL values
null_count = data2.isnull().sum()
print(null_count)

# Lets count NULL percentage
total_data=np.product(data2.shape)
total_null=null_count.sum()
per_of_missing_data=(total_null/total_data)* 100
print('Null percentage:',per_of_missing_data,'%')

sizes = [total_data, total_null]
labels = 'Total data', 'Null data ~2%'
plt.pie(sizes, labels=labels)
plt.show
```

```python
# Find and removes duplicate values and ensuring data cleanliness.

data2.drop_duplicates()
# Ei dublicate-rivejä datassa.

```

```python
ageunder18 = data2.Age[data2.Age < 18].count()
age18to21 = data2.Age[(data2.Age >= 18) & (data2.Age <= 21)].count()
age22to25 = data2.Age[(data2.Age >= 22) & (data2.Age <= 25)].count()
age26to29 = data2.Age[(data2.Age >= 26) & (data2.Age <= 29)].count()
age30to33 = data2.Age[(data2.Age >= 30) & (data2.Age <= 33)].count()
age34to37 = data2.Age[(data2.Age >= 34) & (data2.Age <= 37)].count()
age38to45 = data2.Age[(data2.Age >= 38) & (data2.Age <= 45)].count()
age46to100 = data2.Age[(data2.Age >= 46) & (data2.Age <= 100)].count()


ageunder18pct = (ageunder18 / num_rows) * 100
age18to21pct = (age18to21 / num_rows) * 100
age22to25pct = (age22to25 / num_rows) * 100
age26to29pct = (age26to29 / num_rows) * 100
age30to33pct = (age30to33 / num_rows) * 100
age34to37pct = (age34to37 / num_rows) * 100
age38to45pct = (age38to45 / num_rows) * 100
age46to100pct = (age46to100 / num_rows) * 100


plt.bar(['<18', '18-21', '22-25', '26-29', '30-33', '34-37', '38-45', '46-'], [ageunder18pct, age18to21pct, age22to25pct, age26to29pct, age30to33pct, age34to37pct, age38to45pct, age46to100pct])
plt.ylabel('Percentage')
plt.xlabel('Age')
plt.title('Age distribution')
plt.show()
```

```python
# Rajataan ikäluokkiin -18, 22-25, 26-29, 30-33, 34-37, 38-45 ja 46- ja lisätään taulukkoon
age_ranges = {
    'ageunder18': data2.Age < 18,
    'age18to21': (data2.Age >= 18) & (data2.Age <= 21),
    'age22to25': (data2.Age >= 22) & (data2.Age <= 25),
    'age26to29': (data2.Age >= 26) & (data2.Age <= 29),
    'age30to33': (data2.Age >= 30) & (data2.Age <= 33),
    'age34to37': (data2.Age >= 34) & (data2.Age <= 37),
    'age38to45': (data2.Age >= 38) & (data2.Age <= 45),
    'age46to100': (data2.Age >= 46) & (data2.Age <= 100)
}

# Muutetaan prosenteiksi
def calculate_age_stats(age_range):
    count = data2.Age[age_range].count()
    percentage = (count / num_rows) * 100
    return count, percentage

# Tulostetaan kaavio
plt.bar(['-18', '18-21', '22-25', '26-29', '30-33', '34-37', '38-45', '46-'], [ageunder18pct, age18to21pct, age22to25pct, age26to29pct, age30to33pct, age34to37pct, age38to45pct, age46to100pct])
plt.ylabel('Percentage')
plt.xlabel('Age')
plt.title('Age distribution')
plt.show()
```

```python
nationality_avg = data2.groupby('Nationality')['Name'].count().sort_values(ascending=False)
nationality_avg = nationality_avg[nationality_avg >= 100]  # Exclude nationalities with less than 100 names
nationality_avg.plot(kind='bar')
plt.xlabel('Nationality')
plt.ylabel('Average Number of Players')
plt.title('Nationality with the Most Players on Average')
plt.show()
```

```python
nationality_pct = data2['Nationality'].value_counts(normalize=True)
nationality_pct = nationality_pct[nationality_pct >= 0.019]
nationality_pct = nationality_pct * 100 
nationality_pct.plot(kind='bar')

plt.xlabel('Nationality')
plt.ylabel('Percentage')
plt.title('Nationality percentage of total players')
plt.show()
```
